## Set up

```python
# load the packages
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL of the site
base_site = "https://mydramalist.com/list/1zEbzRK4"
```

```python
# sending a request to the webpage
response = requests.get(base_site)
response.status_code
```

```python
# get the HTML from the webpage
html = response.content
```

```python
# convert the HTML to a BeatifulSoup object
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
# Find all div tags on the webpage containing the information we want to scrape
divs = soup.find_all("div", {"class": "col-xs-10 content"})
print(divs[0])
```

```python
divs[0].contents[1].contents
```

```python
# Extracting all 'h2' tags
headings = [div.find("h2") for div in divs]
headings[0]
# Extracting all titles
titles = [h.text for h in headings]
titles
```

## Extracting the scores

```python
# Filtering only the spans containing the score
spans = soup.find_all("span" , {"class" : "p-l-sm text-muted text-xs"})
len (spans)
```

```python
# Extracting the score string
score = [s.string for s in spans]
score[0]
```

```python
# Removing the '%' sign
score_new = [s[0:6] + s[13:-7] for s in score]
score_new = [s[17:-7] for s in score]
score_new[5]
```

```python
# Converting each score to an integer
int_scr = [int(s) for s in score_new]
int(score_new[0])
new = []
for s in score_new :
    new.extend(int(s))
```
